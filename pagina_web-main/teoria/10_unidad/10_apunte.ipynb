## Variables

Una variable es un espacio donde guardamos información para usarla después.

En Python, para crear una variable usamos el signo `=`. Por ejemplo:

```python

```

Esto significa que estamos guardando el texto `"Ana"` dentro de la variable llamada `nombre`.

A diferencia de R, donde muchas veces usamos `<-`, en Python normalmente usamos `=` para asignar valores.

````python
nombre = "Ana"
edad = 20

print(nombre)
print(edad)
````

# Parte 3 — Tipos de datos y operaciones básicas

Python puede trabajar con diferentes tipos de datos. Algunos de los más comunes son:

- `int`: números enteros, por ejemplo `10`.
- `float`: números decimales, por ejemplo `3.14`.
- `str`: texto o cadenas de caracteres, por ejemplo `"Hola"`.
- `bool`: valores lógicos, es decir `True` o `False`.

También podemos realizar operaciones matemáticas usando operadores.

````python
a = 10
b = 3

print("Suma:", a + b)
print("Resta:", a - b)
print("Multiplicación:", a * b)
print("División:", a / b)
print("Potencia:", a ** b)
````

## Explicación de los operadores

| Operador | Significado | Ejemplo |
|---|---|---|
| `+` | Suma | `a + b` |
| `-` | Resta | `a - b` |
| `*` | Multiplicación | `a * b` |
| `/` | División | `a / b` |
| `**` | Potencia | `a ** b` |

En Python, la potencia se escribe con doble asterisco `**`. En R también se puede usar `^`, pero en Python `^` no se utiliza para elevar potencias.

# Parte 4 — Listas y ciclos

Una lista en Python permite guardar varios valores dentro de una sola variable.

En R trabajamos con vectores, por ejemplo:

```r
numeros <- c(1, 2, 3, 4, 5)
```

En Python usamos corchetes `[]`:

```python
numeros = [1, 2, 3, 4, 5]
```

Una diferencia importante es que Python empieza a contar posiciones desde **0**, no desde 1.

````python
numeros = [1, 2, 3, 4, 5]

print(numeros)
print("Primer elemento:", numeros[0])
print("Último elemento:", numeros[-1])
````

## Ciclo `for`

Un ciclo permite repetir instrucciones.

El ciclo `for` se usa cuando queremos recorrer una secuencia de elementos, como una lista.

En el siguiente ejemplo, Python toma cada elemento de la lista `numeros` y lo guarda temporalmente en la variable `n`.

````python
for n in numeros:
    print(n)
````

# Parte 5 — Introducción a `pandas`

`pandas` es una biblioteca de Python utilizada para trabajar con datos en forma de tablas.

En R, una estructura muy común es el `data.frame`. En Python, con `pandas`, usamos el `DataFrame`.

Un `DataFrame` tiene filas y columnas, similar a una hoja de cálculo.

Para usar `pandas`, primero debemos importarlo.

````python
import pandas as pd
````

## Crear un DataFrame

Primero crearemos un diccionario de Python. Un diccionario guarda información en pares de **clave** y **valor**.

En este caso:

- Las claves serán los nombres de las columnas.
- Los valores serán listas con los datos de cada columna.

````python
datos = {
    "Nombre": ["Ana", "Luis", "Carlos", "María"],
    "Edad": [20, 21, 19, 22],
    "Calificacion": [90, 85, 95, 88]
}
print(datos)
df = pd.DataFrame(datos)

df
````

## Explorar el DataFrame

Cuando trabajamos con datos, es importante revisarlos antes de analizarlos.

Algunas funciones útiles son:

- `head()`: muestra las primeras filas.
- `describe()`: muestra estadísticas básicas.
- Seleccionar una columna usando corchetes, por ejemplo `df["Edad"]`.

````python
df.head()
````

````python
df.describe()
````

````python
df["Edad"]
df["Nombre"]
````

## Calcular estadísticas simples

También podemos calcular valores como el promedio, el valor máximo y el valor mínimo de una columna.

````python
promedio_calificaciones = df["Calificacion"].mean()
maxima_calificacion = df["Calificacion"].max()
minima_calificacion = df["Calificacion"].min()

print("Promedio:", promedio_calificaciones)
print("Calificación máxima:", maxima_calificacion)
print("Calificación mínima:", minima_calificacion)
````

# Parte 6 — Visualización de datos con `matplotlib`

Las gráficas nos ayudan a entender mejor los datos.

En Python podemos usar la biblioteca `matplotlib` para crear visualizaciones básicas.

Primero importamos el módulo `pyplot`, que contiene funciones para crear gráficas.

````python
import matplotlib.pyplot as plt
````

## Gráfica de barras

Una gráfica de barras es útil para comparar valores entre categorías.

En este ejemplo, compararemos las calificaciones de cada estudiante.

````python
plt.bar(df["Nombre"], df["Calificacion"])

plt.title("Calificaciones de estudiantes")
plt.xlabel("Estudiantes")
plt.ylabel("Calificación")

plt.show()
````

## Gráfica de dispersión

Una gráfica de dispersión muestra la relación entre dos variables numéricas.

En este caso, observaremos la relación entre la edad y la calificación.

````python
plt.scatter(df["Edad"], df["Calificacion"])

plt.title("Edad y calificación")
plt.xlabel("Edad")
plt.ylabel("Calificación")

plt.show()
````

# Comparación rápida: R vs Python

| Concepto | R | Python |
|---|---|---|
| Asignación | `x <- 10` | `x = 10` |
| Mostrar resultado | `print(x)` | `print(x)` |
| Vector/lista | `c(1, 2, 3)` | `[1, 2, 3]` |
| Data frame | `data.frame()` | `pd.DataFrame()` |
| Resumen estadístico | `summary()` | `describe()` |
| Leer CSV | `read.csv()` | `pd.read_csv()` |
| Gráficas | `plot()` / `ggplot2` | `matplotlib` |

Aunque R y Python tienen diferencias de sintaxis, ambos permiten analizar datos, crear gráficas y resolver problemas computacionales.
